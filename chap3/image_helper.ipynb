# Image loading and checking the output of CUDA code

```python
from PIL import Image
import numpy as np

image = Image.open('../assets/cat.jpg')
image = image.resize((224, 224))
# save image as binary for cuda code to read
im_arr = np.array(image)
im_arr.tofile('cat.bin')
image
```

## Compile and run cuda code

```python
!nvcc grayscale.cu
!./a.out
```

## Grayscaled output

```python
res = np.fromfile('gray.bin', dtype=np.uint8)
res = res.reshape(224, 224)
Image.fromarray(res)
```

## Gaussian blur of grayscale image

```python
res = np.fromfile('blur.bin', dtype=np.uint8)
res = res.reshape(224, 224)
Image.fromarray(res)
```
